### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
Range = df["Sepal Length (in cm)"].max() - df["Sepal Length (in cm)"].min()
print( "Sepal Length (in cm)  Range ", round(Range,3))

Range = df["Sepal Width in (cm)"].max() - df["Sepal Width in (cm)"].min()
print( " Sepal Width in (cm) Range ", round(Range,3))

Range = df["Petal length (in cm)"].max() - df["Petal length (in cm)"].min()
print( " Petal length (in cm Range ", round(Range,3))


Range = df["Petal width (in cm)"].max() - df["Petal width (in cm)"].min()
print( "Petal width (in cm) Range ", round(Range,3))
```
